```python
import matplotlib.pyplot as plt

type_counts = data['gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors = ['#ff9999', '#66ffcc','#66b3ff']
)
plt.title("Distribution of Content Type")
plt.show()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if 'age' in data.columns:
    sns.histplot(data['age'], bins=30, kde=True)
    plt.xlabel('Age')
    plt.title('Age Distribution')
    plt.show()
```

```python
import plotly.express as px
if 'age' in data.columns:
    fig = px.histogram(df, x='age', nbins=10, title='Age Distribution')
    fig.update_layout(bargap=0.2)
    fig.show()
```

```python
gender_counts = df['gender'].value_counts().reset_index()
gender_counts.columns = ['gender', 'count']

fig = px.bar(gender_counts,
             x='gender', y='count',
             labels={'gender': 'Gender', 'count': 'Count'},
             title='Gender Distribution')
fig.show()
```

```python
if 'age' in data.columns and 'gender' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='gender', y='age', data=data)
    plt.title('Age Distribution by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Age')
    plt.show()
```

```python
if 'age' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data['age'], fill=True, color='skyblue')
    plt.title('Age Density Plot')
    plt.xlabel('Age')
    plt.ylabel('Density')
    plt.show()
```
