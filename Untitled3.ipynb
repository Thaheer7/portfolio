```python
df=pd.DataFrame(data)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if 'age' in data.columns:
    sns.histplot(data['age'], bins=30, kde=True)
    plt.xlabel('Age')
    plt.title('Age Distribution')
    plt.show()
```

```python
import plotly.express as px
if 'age' in data.columns:
    fig = px.histogram(df, x='age', nbins=10, title='Age Distribution')
    fig.update_layout(bargap=0.2)
    fig.show()
```

```python
gender_counts = df['gender'].value_counts().reset_index()
gender_counts.columns = ['gender', 'count']

fig = px.bar(gender_counts,
             x='gender', y='count',
             labels={'gender': 'Gender', 'count': 'Count'},
             title='Gender Distribution')
fig.show()
```

```python
if 'age' in data.columns and 'gender' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='gender', y='age', data=data)
    plt.title('Age Distribution by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Age')
    plt.show()
```

```python
if 'age' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data['age'], fill=True, color='skyblue')
    plt.title('Age Density Plot')
    plt.xlabel('Age')
    plt.ylabel('Density')
    plt.show()
```

```python
top_diagnoses = data['diagnosis_code'].value_counts().head(10)
diagnosis_df = top_diagnoses.reset_index()
diagnosis_df.columns = ['diagnosis', 'count']

plt.figure(figsize=(10, 6))
sns.barplot(data=diagnosis_df,x='count',y='diagnosis',y='diagnosis',dodge=False,palette='viridis',legend=False)
plt.title('Top 10 Most Common Diagnoses')
plt.xlabel('Count')
plt.ylabel('Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', hue='gender', multiple='stack', kde=True)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
top_doctors = data['doctor_id'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_doctors.index.astype(str), y=top_doctors.values,color='yellow')
plt.title('Top 10 Doctors by Visit Count')
plt.xlabel('Doctor ID')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)
plt.show()
```

```python
top_diagnoses = data['diagnosis_code'].value_counts().head(5).index
subset = data[data['diagnosis_code'].isin(top_diagnoses)]

pivot_table = pd.pivot_table(subset, index='diagnosis_code', columns='gender', aggfunc='size', fill_value=0)

sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap of Gender vs Top Diagnoses')
plt.show()
```
